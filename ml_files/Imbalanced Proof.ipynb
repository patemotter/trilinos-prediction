```python
import numpy as np
import pandas as pd
import seaborn as sns
import timeit

c_3_1 = pd.read_csv('./classifications/classified_3_1-12-24.csv', header=0, delimiter=',')
```

```python
ax = sns.countplot(x='numprocs', data=c_3_1, hue='good_or_bad')
for p in ax.patches:
    ax.annotate(int(p.get_height()), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
sns.plt.show()
```
